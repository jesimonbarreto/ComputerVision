```python
%matplotlib inline

import cv2
import numpy as np
from matplotlib import pyplot as plt
```

## Naive Calibration Method

O Naive é baseado na aproximação da distância focal baseado em imagens de objetos não deformados e com dimensões conhecidas. Além dessas informações, é necessário ter a distância desse objeto até a lente de captura.
As fórmulas desse método são:
- fx = (dx/dX) * dZ
- fy = (dy/dY) * dZ

Onde:

dX - Tamanho do objeto no eixo X (em cm)

dY - Tamanho do objeto no eixo Y (em cm)

dZ - Distância do objeto até a câmera (em cm)

dx - tamanho do objeto no eixo X na imagem gerada (em pixels)

dy - tamanho do objeto no eixo Y na imagem gerada (em pixels)


```python
def naive_calibration(dim_real, dim_image):
    #fx = (dx/dX) * dZ
    fx = (dim_image[0]/dim_real[0]) * dim_real[2]
    #fy = (dy/dY)*dZ
    fy = (dim_image[1]/dim_real[1]) * dim_real[2]
    
    focal_p = [fx,fy]
    return focal_p


#Foto do celular no modo normal

print('Normal Image: ')
image_normal = cv2.imread("book_1.jpg")
image_normal = cv2.cvtColor(image_normal, cv2.COLOR_RGB2BGR)


#dim Book in real world and distance (in cm)
dX=28
dY=21.5
dZ=53

dim_real = [dX,dY,dZ]

#dim in image (normal)
dx= 2772 - 1469
dy= 2527 - 836

dim_image_normal = [dx,dy]

focal_normal = naive_calibration(dim_real, dim_image_normal)

print("Focal: X: {}; Y: {}".format(focal_normal[0], focal_normal[1]))

plt.imshow(image_normal)
plt.show()
```

```python
#Foto do celular no modo wide

print('\nWide Image:')

image_wide = cv2.imread("book_2.jpg")
image_wide = cv2.cvtColor(image_wide, cv2.COLOR_RGB2BGR)


#dim in image (wide)
dx= 1504 - 1036
dy= 1274 - 760

dim_image_wide = [dx,dy]

focal_wide = naive_calibration(dim_real, dim_image_wide)

print("Focal: X: {}; Y: {}".format(focal_wide[0], focal_wide[1]))

plt.imshow(image_wide)
plt.show()
```

## Método SVD 

Método implementado com base no livro de introdução a técnicas de computação visual [1].

- Inicialmente foi feita a captura das informações dos pontos da imagem que representam as quinas. Depois, gerados os dados de referência no mundo real. No final, esses dados foram colocados no algoritmo e gerada a resposta esperada.

### Coleta de informações sobre a captura

```python
#Imagem de Referência
image = cv2.imread("alvoD8.jpg")
image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)

#informações retiradas
#pontos da folha na parede começando pelo quadrado (3,3)
img_points_wall = [[1156,1468],[1344,1476], [1524,1480], [1712,1480], [1908,1476], [2088,1472],
              [1184,1620],[1360,1624], [1536,1620], [1716,1624], [1896,1616], [2080,1620],
              [1196,1756],[1376,1752], [1544,1756], [1716,1760], [1888,1752], [2068,1756],
              [1216,1880],[1388,1884], [1540,1880], [1716,1880], [1876,1880], [2046,1876]]

#pontos da folha na mesa começando pelo quadrado (3,2)
img_points_table = [[1212,2104],[1376,2108], [1536,2112], [1704,2108], [1864,2104], [2032,2116],
              [1196,2236],[1360,2232], [1532,2240], [1700,2232], [1868,2240], [2044,2240],
              [1180,2352],[1340,2360], [1520,2360], [1700,2360], [1884,2360], [2060,2380]]

#infos (numero de quadrados no x e no y e tamanho de cada quadrado em mm)
n_poinst_x = 5
n_points_y = 7
size_rec = 30


plt.imshow(image)
plt.show()
```

```python
#Gerando valores de referencia no mundo
def generate_points_world(x, y, size):
    world_points = []
    z = y - 1
    for i in range(x-2):
        for j in range(y-1):
            world_points.append([size*i, -size*j, 0.0])

    for i in range(x-1):
        for k in range(z):
            world_points.append([size*3, size*i, size*(1+k)])
            
    return world_points


def method_SVD(img_points, world_points):
    
    mtx_A = []

    for i, pts in enumerate(zip(img_points, world_points)):
        mtx_A.append([pts[0][0]*pts[1][0], pts[0][0]*pts[1][1], pts[0][0]*pts[1][2], pts[0][0], 
                       -pts[0][1]*pts[1][0], -pts[0][1]*pts[1][1], -pts[0][1]*pts[1][2], -pts[0][1]])
    mtx_A = np.array(mtx_A)
    
    u, d, vt = np.linalg.svd(mtx_A.copy())
    rank_mtx_A = np.linalg.matrix_rank(mtx_A.copy())
    v = vt[-1,:]
    
    #calcula parametros livro[1]
    gamma = np.sqrt(v[0]**2 + v[1]**2 + v[2]**2)
    alpha = np.sqrt(v[4]**2 + v[5]**2 + v[6]**2)
    alpha = alpha/gamma
    v_h = v/gamma
    
    matrix_rot = np.zeros((3,3))
    matrix_rot[0,:] = v_h[-4:-1]/(alpha)
    matrix_rot[1,:] = v_h[:3]
    matrix_rot[2,:] = np.cross(matrix_rot[0,:], matrix_rot[1,:])
    
    matrix_T = np.zeros((3,1))
    matrix_T[:, 0] = [v_h[7]/alpha, v_h[3], 0]
    
    ur, dr, vtr = np.linalg.svd(matrix_rot)
    matrix_rot = np.dot(np.dot(ur,np.identity(3)), vtr)
    
    if (img_points[15][0] * (matrix_rot[0][0]*world_points[15][0] + matrix_rot[0][1]*world_points[15][1] + matrix_rot[0][2]*world_points[15][2] + matrix_T[0]) > 0):
        matrix_rot[0,:] *= -1
        matrix_rot[1,:] *= -1
        matrix_T *= -1
    
    new_mtx_A = np.zeros((len(img_points),2))
    vec_b = np.zeros((len(img_points),1))
    for i, pts in enumerate(zip(img_points, world_points)):
        new_mtx_A[i] = [pts[0][0], matrix_rot[0][0]* pts[1][0] + matrix_rot[0][1]*pts[1][1] + 
                        matrix_rot[0][2]*pts[1][2] + matrix_T[0]]
        vec_b[i] = -pts[0][0]*(matrix_rot[2][0]*pts[1][0] + matrix_rot[2][1]*pts[1][1] + 
                               matrix_rot[2][2]*pts[1][2])
        
    Tz, focal_x = np.dot(np.dot(np.linalg.inv(np.dot(np.transpose(new_mtx_A),new_mtx_A)),np.transpose(new_mtx_A)),vec_b)
    focal_y = focal_x/alpha
    matrix_T[2] = Tz
    
    return [rank_mtx_A, matrix_rot, matrix_T, alpha, gamma, focal_x, focal_y]




world_points = generate_points_world(n_poinst_x, n_points_y, size_rec)



img_points = img_points_wall + img_points_table

param = method_SVD(img_points, world_points)
print('Using SVD')
print('matriz A: {}'.format(param[0]))
print('Alfa: {}\nGama: {}'.format(param[3],param[4]))
print('Matriz de Translação:\n {}'.format(param[2]))
print('Matriz de Rotação:\n {}'.format(param[1]))
print('Focal: X: {}; Y: {}'.format(param[5][0],param[6][0]))
```
